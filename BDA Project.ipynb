```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("creditcard_2023.csv.zip")
df
```

```python
print("Shape of dataset:", df.shape)
print("\nColumns:\n", df.columns)
print("\nInfo:\n")
print(df.info())
```

```python
print(df.shape)
print(df.columns.tolist())
df.head()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Amount'], bins=50, color="skyblue", edgecolor="black")
plt.title("Distribution of Transaction Amount")
plt.xlabel("Amount")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['Class'].value_counts().plot.pie(autopct="%1.1f%%", labels=["Non-Fraud","Fraud"], startangle=90)
plt.title("Fraud vs Non-Fraud Transactions")
plt.show()
```

```python
avg_amount = df.groupby("Class")["Amount"].mean()
plt.figure(figsize=(8,5))
avg_amount.plot(kind="bar", color="purple", edgecolor="black")
plt.title("Average Transaction Amount (Fraud vs Non-Fraud)")
plt.xlabel("Class (0=Non-Fraud, 1=Fraud)")
plt.ylabel("Average Amount")
plt.show()
```

```python
class_count = df["Class"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(class_count, labels=["Non-Fraud","Fraud"], autopct="%1.1f%%", startangle=90)
plt.title("Transaction Count by Class")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['V1'], df['Amount'], alpha=0.5, c=df['Class'], cmap="coolwarm")
plt.title("V1 vs Amount (Color = Fraud)")
plt.xlabel("V1")
plt.ylabel("Amount")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.boxplot(column="Amount", by="Class", grid=False, showfliers=False)
plt.title("Amount Distribution by Class")
plt.suptitle("")  
plt.xlabel("Class (0=Non-Fraud, 1=Fraud)")
plt.ylabel("Amount")
plt.show()
```

```python
import zipfile
import os
zip_file_path = "C:/Users/Dell/Downloads/creditcard_2023.csv.zip"
extract_dir = os.path.dirname(zip_file_path)
try:
    with zipfile.ZipFile(zip_file_path, 'r') as zf:
        zf.extractall(extract_dir)
    print(f"Successfully extracted content from: {zip_file_path}")
    print(f"File(s) placed in: {extract_dir}")
    unzipped_file_path = os.path.join(extract_dir, 'creditcard_2023.csv')
    print(f"New CSV path is: {unzipped_file_path.replace(os.sep, '/')}")
except FileNotFoundError:
    print(f"Error: ZIP file not found at {zip_file_path}. Please check the path.")
except zipfile.BadZipFile:
    print("Error: The file is not a valid ZIP file or is corrupted.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("CreditCardFraudAnalysis").getOrCreate()
file_path = "file:///C:/Users/Dell/Downloads/creditcard_2023.csv" 
df_spark = spark.read.option("header", True).option("inferSchema", True).csv(file_path)
print("DataFrame Schema:")
df_spark.printSchema() 
print("\nFirst 5 rows:")
df_spark.show(5)
print("Distinct classes:", df_spark.select("Class").distinct().count())
```

```python
print("Total transactions:", df_spark.count())
```

```python
print("Transactions in each class:")
df_spark.groupBy("Class").count().show()
```

```python
print("Min and Max transaction amounts:")
df_spark.agg(min("Amount").alias("min_amount"),
             max("Amount").alias("max_amount")).show()
```

```python
print("Average transaction amount per class:")
df_spark.groupBy("Class").agg(avg("Amount").alias("avg_amount")).show()
```

```python
avg_amount = df_spark.agg(avg("Amount").alias("avg")).collect()[0]['avg']
print("Fraudulent transactions with Amount > Average:")
df_spark.filter((col("Class") == 1) & (col("Amount") > avg_amount)) \
  .select("V1", "Amount", "Class").show(10)
```
